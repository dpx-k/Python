```python
import pandas as pd 
```

## Filtering Data

```python
bios = pd.read_csv("datasets/bios.csv")
bios.info()
```

```python
# Grabbing specific columns and applying conditionals on them 

bios.loc[bios['height_cm'] > 215, ['name', 'height_cm']]

# A short hand syntax for the above 
bios[bios['height_cm'] > 215][['name', 'height_cm']]

# applying multiple conditions
bios[(bios['height_cm'] > 215) & (bios['born_country'] == 'USA')][["name", 'born_country', 'height_cm']]

# string ops - grab columns by string
bios[bios['name'].str.contains("Keith", case=False)] # case parameter ignores the case of the string parameter passed

# use a regex expression in pandas
bios[bios['name'].str.contains("keith|patrick", case = False)]

# more complex conditionals
bios[bios['born_country'].isin(["USA", 'FRA', 'GBR']) & (bios['name'].str.startswith("Keith"))]

# using SQL like queries
bios.query('born_country == "USA" and born_city == "Seattle"')
```
